```python
import pandas as pd
import re
from transformers import pipeline

#load in the dataset
df = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='latin-1', header=None)

#clean the tweets so there isn't things like url's, mentions, hashtags, etc, etc
def clean_tweet(tweet):
    tweet = re.sub(r"http\S+|www\S+", '', tweet, flags=re.MULTILINE)
    tweet = re.sub(r'\@\w+|\#', '', tweet)
    return tweet

#function is used to predict bias
def predict_bias(tweet):
    sentiment = model(tweet)
    return "Bias Detected" if sentiment[0]['label'] == 'NEGATIVE' else "No Bias"

#loads in the sentiment analysis model
model = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')

#loop to prompt the user for a topic and breaks if the topic entered isn't valid
while True:
    topic = input("Enter a topic to search in the dataset: ")

    df_filtered = df[df[5].str.contains(topic, case=False, na=False)]

    if df_filtered.empty:
        print(f"Topic '{topic}' not found in the dataset. Please try another topic.")
    else:
        break  # Exit the loop if a valid topic is found

#this limits the amount of tweets that is filtered. I limited it to 100 because if it went through the dataset, it would take
#a very very long time.
df_filtered = df_filtered.head(500)

df_filtered.loc[:, 'cleaned_tweet'] = df_filtered[5].apply(clean_tweet)

#used to predict bias in the tweets
df_filtered['bias'] = df_filtered['cleaned_tweet'].apply(predict_bias)

#then displays/outputs 50 results in a table
df_filtered[['cleaned_tweet', 'bias']].head(50)
```
